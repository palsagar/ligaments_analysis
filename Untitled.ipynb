```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
xspace = np.linspace(10,60,6)
y = np.array([5.5,3.8,1.8,2.5,2.3,2.0])
y2 = np.array([1.8,0.4,0.2,0.1,0.1,0.1])
```

```python
plt.plot(xspace,y);
```

```python
def power_law(x,a,b):
    return a * x**(-b)
```

```python
from useful_functions import * 
popt_pl , pcov_pl = fit(xspace,y,power_law)
```

```python
fit_error = np.sqrt(np.diag(pcov_pl))

standard_fit_error = fit_error[1]/np.sqrt(6)

m_fit = popt_pl[1]
bound_1 = m_fit - 1.96 * standard_fit_error
bound_2 = m_fit + 1.96 * standard_fit_error

print("The value of m is {0:.2f}, with a 95% confidence interval of [{1:.2f}-{2:.2f}]\n".format(-m_fit, -bound_1, -bound_2))

new_xspace = np.linspace(10,60, 1000)

plt.figure(figsize=(8,8))
plt.scatter(xspace,y,label="points");
plt.scatter(xspace,y2,label="points", color='red');
plt.plot(new_xspace, power_law(new_xspace, *popt_pl), linestyle = '--', color = 'black',  
         label="linear fit : $n = %.2f \, , c = %.2f$" %(popt_pl[1], -popt_pl[0]));
#plt.xlabel(r'$\rm{log}(D/W)$');
#plt.ylabel(r'$\rm{log}(\rm{PDF})$');
plt.yscale('log',basey=10);
plt.xscale('log', basex=10);
plt.legend();
```
